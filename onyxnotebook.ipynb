### ONYXDATADNA CHALLENGE JULY 2026

### **1. Business Understanding**
#### **Project Objective**

The objective of this project is to analyze the impact of artificial intelligence (AI) adoption on the global workforce between 2021 and 2024. Using workforce data across countries, industries, and skill categories, the analysis aims to identify trends in AI adoption, evaluate changes in employment and workforce readiness, and examine how the emergence of generative AI in late 2022 has influenced labour market dynamics. The insights generated will support governments and employers in making informed, data-driven decisions on workforce planning, skills development, and AI adoption strategies.

### **1.1 Business Questions**



#### 1. Temporal Trends

1. How has the AI adoption rate changed quarter-over-quarter since the emergence of generative AI in Q4 2022?
2. Is the rate of job displacement accelerating or stabilizing across 2023 and 2024?

#### 2. Industry and Skill Analysis

3. Which industries exhibit high automation susceptibility but low AI investment, indicating under-preparedness for AI transformation?
4. Which skill categories have the largest gap between AI replaceability and the time required for workforce reskilling?

#### 3. Country-Level Analysis

5. Do countries with more mature AI policies achieve better job creation-to-displacement ratios?
6. Which regions show the greatest disparities in AI adoption between developed and emerging economies?

#### 4. Reskilling and Investment

7. Is investment in workforce reskilling keeping pace with the number of jobs displaced by AI across industries?
8. Which country–industry combinations exhibit the largest gap between displacement risk and reskilling investment?

#### 5. Strategic Insights

9. Which countries, industries, and skill categories are best positioned to benefit from AI adoption?
10. What workforce strategies should governments and employers prioritize to minimize displacement while maximizing the benefits of AI adoption?


### **2. Data Understanding**

#### Dataset Overview

| Table Name | Type | Description | Purpose |
|------------|------|-------------|---------|
| `fact_workforce_ai_index` | Fact Table | Contains the measurable AI workforce metrics and links to all dimension tables through foreign keys. | Used for analysis, KPI calculations, and visualizations. |
| `dim_country` | Dimension Table | Stores country-related information such as country name and geographical classifications. | Enables country and regional analysis. |
| `dim_date` | Dimension Table | Contains date attributes such as year, month, and other time-related fields. | Supports trend and time-series analysis. |
| `dim_industry` | Dimension Table | Contains industry names and classifications. | Enables comparison of AI workforce metrics across industries. |
| `dim_skill_category` | Dimension Table | Stores information about AI skill categories. | Supports analysis of workforce skills and skill demand. |
